```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
wcattendance = pd.read_csv("world_cup.csv")
```

```python
wcattendance.head()
```

```python
world_cup_clean = wcattendance.drop(columns=["Host", "Champion", "Runner-Up", "TopScorrer", "Attendance", "Matches"])
world_cup_clean
```

```python
fifa_ranking = pd.read_csv("fifa_ranking_2022_10_06.csv")
```

```python
fifa_ranking.head()
```

```python
# setting up connection to postgres db
connection_string = "postgres:postgres@localhost:5432/Project3"
engine = create_engine(f'postgresql://{connection_string}')

# Confirm tables
engine.table_names()
```

```python
# load 1st of 3 DataFrames into postgres database
world_cup_clean.to_sql(name='world_cup', con=engine, if_exists='append', index=True)
```

```python
# load 2nd of 3 DataFrames into postgres database
fifa_ranking.to_sql(name='fifa_ranking_2022_10_06', con=engine, if_exists='append', index=True)
```
